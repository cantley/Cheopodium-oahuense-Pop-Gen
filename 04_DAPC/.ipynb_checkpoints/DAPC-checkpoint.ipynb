```r
# install packages and dependencies
library(vcfR)
library(adegenet)
install.packages("devtools")
devtools::install_github("zkamvar/ggcompoplot")
library('ggcompoplot')
library('ggplot2')
install.packages("svglite") # to save GGplots as SVG, the superior format
```

```r
# Colorblind-friendly color palette
colorBlind7 <- c("#E69F00", "#56B4E9", "#009E73", "#0072B2", "#D55E00", "#CC79A7")
```

```r
# this script presents the cluster and PCA graphs. This is interactive in an R markdown, but we are limited by running it on a VM/SSH cluster.
inspect_data <- function(vcf_file, max_clust) {
    genlight <- (vcfR2genlight(read.vcfR(vcf_file, verbose = FALSE)))
    find.clusters(genlight, max.n.clust = max_clust) # n-1 will show all possible cluster values
    }

# this script makes the genlight object based on the results above.
make_clusters <- function(vcf_file, n_clust) {
    genlight <- (vcfR2genlight(read.vcfR(vcf_file, verbose = FALSE)))
    n <- nInd(genlight)-1
    find.clusters(genlight, max.n.clust=n, n.clust=n_clust, n.pca=nInd(genlight)) # this retains all the PCAs (we can use fewer)
    }

# this script makes the plots.
scatter_compo <- function(clusters, vcf_file, n_pca, n_da, eig_pos, pca_pos, 
                          comp_posi, n_clusters, pop_names, palette) {
    grp <- clusters
    x   <- vcfR2genlight(read.vcfR(vcf_file, verbose = FALSE))
    dpc <- dapc(x, pop = grp$grp, n.pca=n_pca, n.da=n_da)
    scatterplot_handle <- scatter(dpc, posi.da = eig_pos, scree.pca=TRUE, posi.pca = pca_pos, col=palette)
    scatterplot_handle
    assignplot_handle <- assignplot(dpc, subset=1:nInd(x))
    assignplot_handle
    compoplot_handle <- compoplot(dpc, posi=comp_posi, txt.leg=paste("Cluster",1:n_clusters), lab="", n.col=1, 
          xlab="Individuals", col=palette)
    compoplot_handle
    optim.a.score(dpc)
    
    svg(paste0(basename(vcf_file),".scatterplot.svg"), width=3.5, height=3)
    scatterplot_handle
    dev.off()

    svg(paste0(basename(vcf_file),".assignplot.svg"))
    assignplot_handle
    dev.off()
    
    svg(paste0(basename(vcf_file),".compoplot.svg"))
    compoplot_handle
    dev.off()
    }

```

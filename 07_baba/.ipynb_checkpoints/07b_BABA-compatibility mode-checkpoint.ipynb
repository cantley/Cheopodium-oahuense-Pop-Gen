```python
## ipyrad and raxml output files. You'll need to set these
loci_ALL = LOCI_PATH+"/chenopodium_20250304b-tetraploid.indivmiss60.locimiss50.thin10.maf0.01.LD0.8-1000.loci"
loci_OAM = LOCI_PATH+"/chenopodium_20250304b-tetraploid.indivmiss60.locimiss50.thin10.maf0.01.LD0.8-1000.loci"
loci_OAM_OAout = LOCI_PATH+"/chenopodium_20250304b-tetraploid.indivmiss60.locimiss50.thin10.maf0.01.LD0.8-1000.no-AR.no-CH-except-OA.loci"

newk_ALL = "/home/923643692/cheno_pop_tests/06_phylos/analysis-raxml/RAxML_bestTree.chenopodium_20250304b-tetraploid.indivmiss60.locimiss50.thin10.maf0.01.LD0.8-1000.min4.raxml.renamed"
newk_OAM = "/home/923643692/cheno_pop_tests/06_phylos/analysis-iqtree/chenopodium_20250304b-tetraploid.indivmiss60.locimiss50.thin10.maf0.01.LD0.8-1000.no-AR.no-CH-except-OA.min4.phy.contree"
newk_OAM_OAout = "/home/923643692/cheno_pop_tests/06_phylos/analysis-iqtree/chenopodium_20250304b-tetraploid.indivmiss60.locimiss50.thin10.maf0.01.LD0.8-1000.no-AR.no-CH-except-OA.min4.phy.contree"
```

```python
## parse the newick tree, re-root it, and plot it.
rtre = toytree.tree(newk_OAM_OAout).root(wildcard="CH")
rtre.draw(
    height = 800,
    width  = 800,
#    node_labels=rtre.get_node_values("support")
    )

## store rooted tree back into a newick string.
newick = rtre.write()
newick
```

```python
## create a baba object linked to a data file and newick tree
bb = ipa.baba(data=loci_OAM_OAout, newick=newick)
```

```python
pops = ["CH", "IP", "MO", "PB", "PK"]

match_dict = {}
for pop in pops:
    match_dict[pop] = []

pattern = "CO-[A-Z]+-[0-9]+"
matches = re.findall(pattern, newick)
for match in matches:
    for pop in pops:
        if "CO-" + pop + "-" in match:
            match_dict[pop].append(match)
print(match_dict['CH'])
```

```python
print(set(bb.samples))
```

```python
## generate all possible abba-baba tests meeting a set of constraints
bb.generate_tests_from_tree(
    constraint_dict={
        "p4": match_dict['CH'],
    }
)
```

```python
bb.run(ipyclient)
```
